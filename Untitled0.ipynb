```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Initialisation des matrices
```

```python
nx = 256
ny = 256
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x, y)

plt.figure()
f, axarr = plt.subplots(1,2)
axarr[0].imshow(xv, cmap='gray')
axarr[1].imshow(yv, cmap='gray')
```

```python
# fonction grille sinusoïdale

def fabriquer_grille_sin(angle, frequence, phase):
    rampe = (np.cos(angle) * xv + np.sin(angle) * yv)
    grille_sin = np.sin(frequence * 2 * np.pi * rampe + phase) / 2 + 0.5
    return grille_sin

plt.imshow(fabriquer_grille_sin(np.pi/4, 6, 0), cmap='gray', vmin = 0, vmax = 1)
```

```python
# fonction cercles sinusoïdaux

def fabriquer_cercles_sin(angle, frequence, phase) : 
  rayons = np.sqrt((xv - 0.5) ** 2 + (yv - 0.5) ** 2)
  cercles_sin = np.sin(frequence * 2 * np.pi * rayons + phase) / 2 + 0.5;
  return cercles_sin
  
plt.imshow(fabriquer_cercles_sin(np.pi/4, 10, 0), cmap = 'gray', vmin = 0, vmax = 1)
```

```python
# fonction secteurs sinusoïdaux

def fabriquer_secteurs_sin(frequence, phase):
  angles = np.arctan2((yv - 0.5), (xv - 0.5))
  secteurs_sin = np.sin(frequence * angles + phase) / 2 + 0.5
  return secteurs_sin

plt.imshow(fabriquer_secteurs_sin(10, 0), cmap = 'gray', vmin = 0, vmax = 1)
```

```python
# fonction "wiggles" sinusoïdaux

def fabriquer_wiggles_sin(freq_min, freq_max, freq_rad, phase_rad, phase):
  angles = np.arctan2((yv - 0.5), (xv - 0.5))
  modulation_freq = (freq_max - freq_min) * (np.sin(freq_rad * angles + phase_rad) / 2 + 0.5) + freq_min
  rayons = np.sqrt((xv - 0.5) ** 2 + (yv - 0.5) ** 2)
  wiggle_sin = np.sin(modulation_freq * 2 * np.pi * rayons + phase) / 2 + 0.5
  return wiggle_sin

plt.imshow(fabriquer_wiggles_sin(7, 10, 6, 0, 0), cmap = 'gray', vmin = 0, vmax = 1)
```

```python
# gaussienne

def fabriquer_gaussienne(ecart_type):
  gaussienne = np.exp(-((xv - 0.5) ** 2 / ecart_type ** 2) - ((yv - 0.5) ** 2 / ecart_type ** 2))
  return gaussienne
  
plt.imshow(fabriquer_gaussienne(0.2), cmap = 'gray', vmin = 0, vmax = 1)
```

```python
# gabor

def fabriquer_gabor(ecart_type, angle, frequence, phase):
  gaussienne = fabriquer_gaussienne(ecart_type)
  grille_sin = fabriquer_grille_sin(angle, frequence, phase)
  gabor = gaussienne * (grille_sin - 0.5) + 0.5
  return gabor

plt.imshow(fabriquer_gabor(0.2, np.pi/4, 6, 0), cmap = 'gray', vmin = 0, vmax = 1)
```

```python
# grille sinusoïdale en couleur (vert et rouge)

def fabriquer_grille_sin_rv(angle, frequence, phase):
  grille_sin = fabriquer_grille_sin(angle, frequence, phase)
  grille_sin_rgb = np.zeros((grille_sin.shape[0], grille_sin.shape[1], 3))
  grille_sin_rgb[:, :, 0] = grille_sin  # rgb : r(red) g(green) b(blue)  canal rouge
  grille_sin_rgb[:, :, 1] = 1 - grille_sin  # donne le complément, dans le canal vert
  return grille_sin_rgb

plt.imshow(fabriquer_grille_sin_rv(np.pi/4, 6, 0), cmap='gray')
```

```python
# bleu et jaune

def fabriquer_grille_sin_jb(angle, frequence, phase):
  grille_sin = fabriquer_grille_sin(angle, frequence, phase)
  grille_sin_rgb = np.zeros((grille_sin.shape[0], grille_sin.shape[1], 3))
  grille_sin_rgb[:, :, 0] = grille_sin 
  grille_sin_rgb[:, :, 1] = grille_sin 
  grille_sin_rgb[:, :, 2] = 1-grille_sin
  return grille_sin_rgb

plt.imshow(fabriquer_grille_sin_jb(np.pi/4, 6, 0), cmap='gray')
```

```python
# grille sinusoïdale verte et rouge avec fenêtre gaussienne

grille_sin_vr = fabriquer_grille_sin_rv(np.pi/4,4,4)
gaussienne = fabriquer_gaussienne(0.2)

plt.imshow(grille_sin_vr, cmap='gray', vmin=0, vmax=1)
plt.imshow(gaussienne, cmap='gray', vmin=-1, vmax=1, alpha= 1 - gaussienne) # -1 à 1 pour avoir du gris
```

```python
# grille sinusoïdale bleu et jaune avec fenêtre gaussienne

grille_sin_jb = fabriquer_grille_sin_jb(np.pi/4,4,4)

plt.imshow(grille_sin_jb, cmap='gray', vmin=0, vmax=1)
plt.imshow(gaussienne, cmap='gray', vmin=-1, vmax=1, alpha= 1 - gaussienne)
```
